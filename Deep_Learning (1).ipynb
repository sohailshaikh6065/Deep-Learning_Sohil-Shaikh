```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from google.colab import files
uploaded = files.upload()
df = pd.read_csv(list(uploaded.keys())[0])

# Sigmoid function
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Function to compute the loss (Log-Loss)
def compute_loss(X, y, theta):
    m = len(y)
    predictions = sigmoid(X.dot(theta))
    loss = -(1 / m) * (y.T.dot(np.log(predictions)) + (1 - y).T.dot(np.log(1 - predictions)))
    return loss[0, 0]

# Function to compute the gradient
def compute_gradient(X, y, theta):
    m = len(y)
    predictions = sigmoid(X.dot(theta))
    gradients = (1 / m) * X.T.dot(predictions - y)
    return gradients

# Gradient Descent function
def gradient_descent(X, y, theta, learning_rate, iterations):
    loss_history = []
    for i in range(iterations):
        gradients = compute_gradient(X, y, theta)  # Compute gradients
        theta -= learning_rate * gradients         # Update parameters
        loss = compute_loss(X, y, theta)          # Compute loss
        loss_history.append(loss)

        # Print loss at intervals
        if i % 100 == 0:
            print(f"Iteration {i}: Loss = {loss:.4f}")
    return theta, loss_history

# Generate sample data for binary classification
np.random.seed(42)
X = np.random.rand(100, 1) * 10  # Features (scaled between 0 and 10)
y = (X > 5).astype(int)          # Binary target (0 or 1)
y = y.reshape(-1, 1)

# Add a bias term (intercept) to X
X_b = np.c_[np.ones((100, 1)), X]  # Add a column of ones to X

# Initialize parameters
theta_initial = np.random.randn(2, 1)  # 2 parameters (intercept and slope)
learning_rate = 0.01
iterations = 1000

# Run Gradient Descent
theta_optimal, loss_history = gradient_descent(X_b, y, theta_initial, learning_rate, iterations)

# Final Results
final_loss = compute_loss(X_b, y, theta_optimal)

# Display Results
print("\nOptimal Parameters (Theta):")
print(theta_optimal)
print(f"\nFinal Loss: {final_loss:.4f}")

# Plot the loss curve
plt.figure(figsize=(8, 5))
plt.plot(range(iterations), loss_history, label="Loss Curve")
plt.title("Loss Curve During Gradient Descent")
plt.xlabel("Iterations")
plt.ylabel("Loss (Log-Loss)")
plt.legend()
plt.grid(True)
plt.show()

# Plot the decision boundary
plt.figure(figsize=(8, 5))
plt.scatter(X, y, color="blue", label="Data Points")
x_values = np.linspace(0, 10, 100)
y_values = sigmoid(np.c_[np.ones((100, 1)), x_values].dot(theta_optimal))
plt.plot(x_values, y_values, color="red", label="Decision Boundary")
plt.axhline(0.5, color="green", linestyle="--", label="0.5 Threshold")
plt.title("Logistic Regression Decision Boundary")
plt.xlabel("X (Feature)")
plt.ylabel("Probability")
plt.legend()
plt.grid(True)
plt.show()
```

```python
def mae(y_predicted, y_true):
    total_error = 0
    for yp, yt in zip(y_predicted, y_true):
        total_error += abs(yp - yt)
    print("Total error is:",total_error)
    mae = total_error/len(y_predicted)
    print("Mean absolute error is:",mae)
    return mae
```
